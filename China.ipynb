```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("GlobalLandTemperaturesByCountry11.csv", index_col=0, header=0,  parse_dates=True, squeeze=True)
china1=df[(df.Country == 'China')]
china=china1.drop("AverageTemperatureUncertainty", axis=1)
data_chi_year1 = china.groupby([china.index.year]).mean()
data_chi_year=data_chi_year1[12:193]
data_smooth=data_chi_year.ewm(com=5).mean()
plt.figure(figsize=(20,5),facecolor=None)
ax=plt.axes()
plt.xlabel('Years', fontsize=15)
plt.ylabel('Average temperature (°C)', fontsize=15)
ax.set_facecolor("#AAAAAA")
plt.plot(data_smooth,"#001f3f",linewidth=2,label="SmoothData")
plt.plot(data_chi_year,"#3D9970" ,linewidth=1,label="AbsoluteData")
plt.legend()
plt.title('China', fontsize=30)
plt.savefig('China', dpi=500)
plt.show()

```
