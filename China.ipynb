```python
import pandas as pd
import matplotlib.pyplot as plt
```

In this notebook file we plotted different countries and then built comparison plots.

```python
df = pd.read_csv("GlobalLandTemperaturesByCountry11.csv", index_col=0, header=0,  parse_dates=True, squeeze=True)
china1=df[(df.Country == 'China')]
china=china1.drop("AverageTemperatureUncertainty", axis=1)
data_chi_year1 = china.groupby([china.index.year]).mean()
data_chi_year=data_chi_year1[12:193]
data_smooth_chi=data_chi_year.ewm(com=5).mean()
plt.figure(figsize=(20,4),facecolor="white")
ax=plt.axes()
plt.xlabel('Years', fontsize=15)
plt.ylabel('Average temperature (C)', fontsize=15)
ax.set_facecolor("#AAAAAA")
plt.plot(data_smooth_chi,"#FF4136",linewidth=2,label="China average")
plt.plot(data_chi_year,"#FF851B" ,linewidth=1,label="China absolute")
plt.legend()
plt.title('China', fontsize=30)
plt.savefig('China', dpi=500)
ax.annotate("Some explanation",xy=(1800,10))
plt.show()

```

```python
df = pd.read_csv("GlobalLandTemperaturesByCountry11.csv", index_col=0, header=0,  parse_dates=True, squeeze=True)
india1=df[(df.Country == 'India')]
india=india1.drop("AverageTemperatureUncertainty", axis=1)
data_ind_year1 = india.groupby([india.index.year]).mean()
data_ind_year=data_ind_year1[70::]
data_smooth_ind=data_ind_year.ewm(com=5).mean()
plt.figure(figsize=(13,3),facecolor="white")
ax=plt.axes()
plt.xlabel('Years', fontsize=15)
plt.ylabel('Average temperature (C)', fontsize=15)
ax.set_facecolor("#AAAAAA")
plt.plot(data_smooth_ind,"#85144b",linewidth=2,label="India smooth")
plt.plot(data_ind_year,"#B10DC9" ,linewidth=1,label="India actual")
plt.legend(fontsize="xx-large")
plt.title('India', fontsize=30)
plt.savefig('India', dpi=500)
plt.show()
```

```python
df = pd.read_csv("GlobalLandTemperaturesByCountry11.csv", index_col=0, header=0,  parse_dates=True, squeeze=True)
azerbaijan=df[(df.Country == 'Azerbaijan')]
aze=azerbaijan.drop("AverageTemperatureUncertainty", axis=1)
data_aze_year1 = aze.groupby([aze.index.year]).mean()
data_aze_year=data_aze_year1[2:205]
data_smooth_aze=data_aze_year.ewm(com=5).mean()
plt.figure(figsize=(20,4),facecolor="white")
ax=plt.axes()
plt.xlabel('Years',fontsize=15)
plt.ylabel('Average temperature (C)',fontsize=15)
ax.set_facecolor("#AAAAAA")
plt.plot(data_smooth_aze,"#001f3f",linewidth=2,label="Azerbaijan average")
plt.plot(data_aze_year,"#3D9970" ,linewidth=1,label="Azerbaijan Absolute")
plt.legend()
plt.title('Azerbaijan', fontsize=30)
plt.savefig('Azerbaijan', dpi=500)
plt.show()

```

```python
#Absolute temperatures, annual readings for Estonia
data_abs_est = pd.read_csv("Estonia.csv",index_col=1, header=0,  parse_dates=True, squeeze=True)
data_est1=data_abs_est.drop(["Unnamed: 0","AverageTemperatureUncertainty","Country"], axis=1)
data_est=data_est1[2:3122]
data_est_year = data_est.groupby([data_est.index.year]).mean()
data_smooth_est=data_est_year.ewm(com=5).mean()
plt.figure(figsize=(20,4),facecolor="white")
ax=plt.axes()
ax.set_facecolor("#AAAAAA")
plt.plot(data_smooth_est,"#3D9970",linewidth=2,label="Estonia average")
plt.plot(data_est_year,"#2ECC40" ,linewidth=1,label="Estonia Absolute")
plt.legend()
plt.xlabel("Year",fontsize=30)
plt.ylabel("Temp. C",fontsize=30)
plt.show()




```
